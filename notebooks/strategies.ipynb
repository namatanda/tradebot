```python
from datetime import datetime
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np 
```

```python
from scipy.stats import linregress
```

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 6)
plt.ioff()
```

```python
file_path = '../data/snp500.csv'
```

```python
df = pd.read_csv(filepath_or_buffer=file_path, index_col='date', parse_dates=True)
```

```python
df = df.loc[:, ~df.columns.duplicated()]
df.head()
```

```python
def momentum(n):
    returns = np.log(n)
    x = np.arange(stop=len(returns))
    slope, intercept, r_value, p_value, std_err = linregress(x, returns)
    return ((1 + slope)**252 * (r_value**2))
```

```python
momentums = df.copy(deep=True)
```

```python
momentums['momentum'] = df['adj_close'].rolling(window=90).apply(momentum, raw=False)
```

```python
momentums.tail()
```

```python
#apply the momentum function to each ticker in the dataframe

tickers = list(set(df['ticker']))

for ticker in tickers:
    momentums.loc[df['ticker'] == ticker, 'momentum'] = df.loc[df['ticker'] == ticker, 'adj_close'].rolling(window=90).apply(momentum, raw=False)
momentums.tail()
```
